```python
import prefab as pf
```

```python
device = pf.read.from_img(
    img_path="cross_16x128_256x256.png", img_width_nm=1000
)
device.plot()
```

```python
MODEL_NAME = "ANT_NanoSOI_ANF1_d9"
prediction = device.predict(model=pf.models[MODEL_NAME])
prediction_bin = prediction.binarize()

prediction.plot()
prediction_bin.plot()
```
